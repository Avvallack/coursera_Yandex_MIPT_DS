```python
import pandas as pd
```

```python
milk = pd.read_csv('monthly-milk-production.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
```

```python
import seaborn as sns
milk.head()
```

```python
sns.lineplot(data=milk)
```

```python
import statsmodels.api as sm
sm.tsa.stattools.adfuller(milk.milk)
```

```python
milk['days'] = milk.index.days_in_month
```

```python
milk['aver'] = milk.milk / milk.days
```

```python
milk.aver.sum()
```

```python
sns.lineplot(milk.index, milk.aver)
```

```python
milk['dif1'] = milk.aver - milk.aver.shift(1)
```

```python
from  statsmodels.tsa.stattools import adfuller
```

```python
adfuller(milk.dif1[1:])
```

```python
milk['dif12'] = milk.aver - milk.aver.shift(12)
```

```python
adfuller(milk.dif12[12:])
```

```python
milk.aver.count()
```

```python
milk['dif13'] = milk.dif12 - milk.dif12.shift(1)
```

```python
adfuller(milk.dif13[13:])
```

```python
import statsmodels.api as sm
sm.graphics.tsa.plot_acf(milk.dif13[13:].values.squeeze(), lags=50)
```

```python
sm.graphics.tsa.plot_pacf(milk.dif13[13:].values.squeeze(), lags=50)
```
